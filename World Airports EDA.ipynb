```python
#Missing values checking
df.isnull().sum()
```

```python
def calculate_null_percentage(df):
    total_rows = len(df)
    null_percentages = {}
    
    for column in df.columns:
        null_count = df[column].isnull().sum()
        null_percentage = (null_count / total_rows) * 100
        null_percentage = round(null_percentage, 2)  # Round to two decimal places
        null_percentages[column] = null_percentage
    
    return null_percentages
```

```python
null_percentages = calculate_null_percentage(df)
# Print the null percentages
for column, percentage in null_percentages.items():
    print(f"'{column}' has {percentage}% nulls.")
```

We can remove the columns with the most missing values.

```python
df.drop(['home_link', 'wikipedia_link', 'keywords', 'continent', 'iata_code'], axis=1, inplace=True)
```

<h1 style="padding: 20px;color:white;margin:10;font-size:90%;text-align:left;display:fill;border-radius:10px;overflow:hidden;background-image: url(https://img.rawpixel.com/s3fs-private/rawpixel_images/website_content/v1016-c-08_1-ksh6mza3.jpg?w=800&dpr=1&fit=default&crop=default&q=65&vib=3&con=3&usm=15&bg=F4F4F3&ixlib=js-2.2.1&s=f584d8501c27c5f649bc2cfce50705c0)"><b><span style='color:white'>✈ EDA and Data Visualizations</span></b> </h1>

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Distribution of Elevation <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['elevation_ft'].dropna(), kde=True, color='skyblue')

plt.title('Distribution of Elevation', fontsize=16)
plt.xlabel('Elevation (ft)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--')

# Add vertical line for mean
mean_elevation = df['elevation_ft'].mean()
plt.axvline(mean_elevation, color='red', linestyle='--', label='Mean Elevation')

plt.legend()
plt.margins(0.02)

plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Number of Airports by Type <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='type', palette='Set2')

plt.title('Number of Airports by Type', fontsize=16)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--')


plt.tight_layout()
plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Percentage of Airports with Scheduled Service <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(8, 6))
df['scheduled_service'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percentage of Airports with Scheduled Service')
plt.ylabel('')
plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Elevation Distribution by Airport Type <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='type', y='elevation_ft')
plt.title('Elevation Distribution by Airport Type')
plt.xlabel('Type')
plt.ylabel('Elevation (ft)')
plt.xticks(rotation=45)
plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Top 10 Countries with the Highest Number of Airports <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
top_countries = df['iso_country'].value_counts().nlargest(10)

plt.figure(figsize=(10, 8))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 10 Countries with the Highest Number of Airports')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Geographical Distribution of Airports/Heliports <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='longitude_deg', y='latitude_deg', hue='type', alpha=0.7)
plt.title('Geographical Distribution of Airports/Heliports')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

<p style="color:#F6F6F6;background-color:#D61C4E;text-align:center;border-radius:10px 10px;font-weight:bold;font-size:22px">Geographical Distribution of Airports/Heliports with Elevation <span style='font-size:28px; background-color:#F6F6F6 ;'></span></p>

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['longitude_deg'], df['latitude_deg'], c=df['elevation_ft'], cmap='viridis', alpha=0.7)
plt.title('Geographical Distribution of Airports/Heliports with Elevation')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(scatter, label='Elevation (ft)')
plt.show()
```
